```python
resized = cv2.resize(al_map.xray_data_scaled, None, fx=2, fy=2, interpolation=cv2.INTER_LINEAR)
plt.imshow(resized)
plt.title("Al map - new size: " + str("2x"))
plt.show()
```

```python
size_100 = (100, 100)
resized_up = cv2.resize(resized,size_100,cv2.INTER_CUBIC)
plt.imshow(resized_up)
plt.title("Al map - new size: " + str(size_100))
plt.show()
```
